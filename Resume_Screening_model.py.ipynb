```python
import pandas as pd
import numpy as np

def select_applicants(dataset_path, criteria):
    # Read the dataset from Excel file
    dataset = pd.read_excel(dataset_path)

    # Convert 'Performance_PG', 'Performance_UG', 'Performance_12', and 'Performance_10' columns to numeric
    dataset['Performance_PG'] = pd.to_numeric(dataset['Performance_PG'], errors='coerce')
    dataset['Performance_UG'] = pd.to_numeric(dataset['Performance_UG'], errors='coerce')
    dataset['Performance_12'] = pd.to_numeric(dataset['Performance_12'], errors='coerce')
    dataset['Performance_10'] = pd.to_numeric(dataset['Performance_10'], errors='coerce')

    # Filter applicants based on criteria
    if np.isnan(criteria['pg_performance']):                      # in case value is NA
        selected_applicants = dataset[(dataset['Python (out of 3)'] >= criteria['python_score']) &
                                      (dataset['Machine Learning (out of 3)'] >= criteria['ml_score']) &
                                      (dataset['Natural Language Processing (NLP) (out of 3)'] >= criteria['nlp_score']) &
                                      (dataset['Deep Learning (out of 3)'] >= criteria['dl_score']) &
                                      ((dataset['Degree'] == 'B.Tech') |
                                       (dataset['Degree'] == 'M.Tech') |
                                       (dataset['Degree'] == 'MSc')) &
                                      (dataset['Current Year Of Graduation'] <= criteria['graduation_year']) &
                                      (dataset['Performance_UG'] >= criteria['ug_performance']) &
                                      (dataset['Performance_12'] >= criteria['grade_12']) &
                                      (dataset['Performance_10'] >= criteria['grade_10'])]

    elif np.isnan(criteria['ug_performance']):                    # in case value is NA
        selected_applicants = dataset[(dataset['Python (out of 3)'] >= criteria['python_score']) &
                                      (dataset['Machine Learning (out of 3)'] >= criteria['ml_score']) &
                                      (dataset['Natural Language Processing (NLP) (out of 3)'] >= criteria['nlp_score']) &
                                      (dataset['Deep Learning (out of 3)'] >= criteria['dl_score']) &
                                      ((dataset['Degree'] == 'B.Tech') |
                                       (dataset['Degree'] == 'M.Tech') |
                                       (dataset['Degree'] == 'MSc')) &
                                      (dataset['Current Year Of Graduation'] <= criteria['graduation_year']) &
                                      (dataset['Performance_PG'] >= criteria['pg_performance']) &
                                      (dataset['Performance_12'] >= criteria['grade_12']) &
                                      (dataset['Performance_10'] >= criteria['grade_10'])]

    elif np.isnan(criteria['grade_12']):                        # in case value is NA
      selected_applicants = dataset[(dataset['Python (out of 3)'] >= criteria['python_score']) &
                                      (dataset['Machine Learning (out of 3)'] >= criteria['ml_score']) &
                                      (dataset['Natural Language Processing (NLP) (out of 3)'] >= criteria['nlp_score']) &
                                      (dataset['Deep Learning (out of 3)'] >= criteria['dl_score']) &
                                      ((dataset['Degree'] == 'B.Tech') |
                                       (dataset['Degree'] == 'M.Tech') |
                                       (dataset['Degree'] == 'MSc')) &
                                      (dataset['Current Year Of Graduation'] <= criteria['graduation_year']) &
                                      (dataset['Performance_PG'] >= criteria['pg_performance']) &
                                      (dataset['Performance_UG'] >= criteria['ug_performance']) &
                                      # (dataset['Performance_12'] >= criteria['grade_12']) &
                                      (dataset['Performance_10'] >= criteria['grade_10'])]

    elif np.isnan(criteria['grade_10']):                        # in case value is NA
      selected_applicants = dataset[(dataset['Python (out of 3)'] >= criteria['python_score']) &
                                      (dataset['Machine Learning (out of 3)'] >= criteria['ml_score']) &
                                      (dataset['Natural Language Processing (NLP) (out of 3)'] >= criteria['nlp_score']) &
                                      (dataset['Deep Learning (out of 3)'] >= criteria['dl_score']) &
                                      ((dataset['Degree'] == 'B.Tech') |
                                       (dataset['Degree'] == 'M.Tech') |
                                       (dataset['Degree'] == 'MSc')) &
                                      (dataset['Current Year Of Graduation'] <= criteria['graduation_year']) &
                                      (dataset['Performance_PG'] >= criteria['pg_performance']) &
                                      (dataset['Performance_UG'] >= criteria['ug_performance']) &
                                      (dataset['Performance_12'] >= criteria['grade_12'])]
                                      # (dataset['Performance_10'] >= criteria['grade_10'])]

    else:
        selected_applicants = dataset[(dataset['Python (out of 3)'] >= criteria['python_score']) &
                                      (dataset['Machine Learning (out of 3)'] >= criteria['ml_score']) &
                                      (dataset['Natural Language Processing (NLP) (out of 3)'] >= criteria['nlp_score']) &
                                      (dataset['Deep Learning (out of 3)'] >= criteria['dl_score']) &
                                      ((dataset['Degree'] == 'B.Tech') |
                                       (dataset['Degree'] == 'M.Tech') |
                                       (dataset['Degree'] == 'MSc')) &
                                      (dataset['Current Year Of Graduation'] <= criteria['graduation_year']) &
                                      (dataset['Performance_PG'] >= criteria['pg_performance']) &
                                      (dataset['Performance_UG'] >= criteria['ug_performance']) &
                                      (dataset['Performance_12'] >= criteria['grade_12']) &
                                      (dataset['Performance_10'] >= criteria['grade_10'])]

    # Return the selected applicants as a pandas DataFrame
    return selected_applicants

# Example usage:
dataset_path = '/content/Applications_for_Machine_Learning_internship_edited.xlsx'

# Define the selection criteria - we can define according to our convinience
criteria = {
    'python_score': 0,
    'ml_score': 0,
    'nlp_score': 0,
    'dl_score': 0,
    'degree': ['B.Tech', 'M.Tech', 'MSc'],
    'graduation_year': 2026,
    'pg_performance': 5.0,
    'ug_performance': 5.0,
    'grade_12': 60,
    'grade_10': 60
}

selected_applicants = select_applicants(dataset_path, criteria)

# Print the selected applicants
print(selected_applicants)
```
